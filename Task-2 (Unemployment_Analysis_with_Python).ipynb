```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading Dataset
dataset = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
dataset.head()
```

```python
#To know the no. of Rows and Columns
dataset.shape
```

```python
#get the information about the dataset
dataset.info()
```

```python
#Check whether there are any null values in columns or not
dataset.isnull().sum()
```

```python
#Here we change the incorrect names of columns to correct 
dataset.columns=["State","Date","Frequency","Estimated_Unemployment_Rate","Estimated_Employed","Estimated_Labour_Participation_Rate","Region","Longitude","Latitude"]
dataset.columns
```

```python
#correlation
plt.figure(figsize=(10,10))
sns.heatmap(dataset.corr())
```

```python
#Histogram to analyse the estimated employement in india acc. to states
plt.figure(figsize =(20,10))
plt.title("Estimated employment in India according to states during COVID-19")
h = sns.histplot(x = "Estimated_Employed", hue = "State", data = dataset)
sns.move_legend(h, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
#Histogram to analyze the estimated employement in india acc. to region
plt.figure(figsize=(20,10))
plt.title("Estimated Employed in India according to region during COVID-19")
sns.histplot(x = "Estimated_Employed", hue = "Region", data = dataset)
plt.show()
```

```python
#Histogram to analyze the Unemployment rate in India according to states during COVID-19
plt.figure(figsize =(12,10))
plt.title("Unemployment rate in India according to states during COVID-19")
h = sns.histplot(x = "Estimated_Unemployment_Rate", hue = "State", data = dataset)
sns.move_legend(h, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
#Histogram to analyze the Unemployment rate in India according to region during COVID-19
plt.figure(figsize =(12,10))
plt.title("Unemployment rate in India according to region during COVID-19")
sns.histplot(x = "Estimated_Unemployment_Rate", hue = "Region", data = dataset)
plt.show()
```

```python
#piechart to show the Unemployment rate in India according to States during COVID-19
plt.figure(figsize =(12,10))
su = dataset.groupby(dataset['State'])['Estimated_Unemployment_Rate'].sum()
plt.axis('equal')
plt.title("Unemployment rate in India according to States during COVID-19")
plt.pie(su,labels=su.index)
plt.show()
```

```python
#Piechart to show the Unemployment rate in India according to Regions during COVID-19
plt.figure(figsize =(11,6))
su = dataset.groupby(dataset['Region'])['Estimated_Unemployment_Rate'].sum()
plt.axis('equal')
plt.title("Unemployment rate in India according to Regions during COVID-19")
plt.pie(su,labels=su.index)
plt.show()
```
